```python
# Cargamos el archivo csv
df = pd.read_csv('/content/Detalle precios y productos fabricados 2022.csv')
df
```

```python
#Filtro por objeto
#Obtenemos todos los registros de la base en donde el nombre sea
# LETICIA RAMIREZ HERNANDEZ
filtro1=df[df["NOMBRE"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1.to_csv("filtro1.csv")
files.download("filtro1.csv")
filtro1
```

```python
#filtro por filas
# Obtenemos la filas 7, 9 y 15
filtro2=df.iloc[[7,9,15],:]
filtro2.to_csv("filtro2.csv")
files.download("filtro2.csv")
filtro2
```

```python
# Obtenemos las columnas 1, 3 y 5(Fecha_DOC, NOMBRE.1, CVE_ART	)
filtro3=df.iloc[: , [1,3,5]]
filtro3.to_csv("filtro3.csv")
files.download("filtro3.csv")
filtro3
```

```python
df1 = pd.read_csv('/content/Detalle precios y productos fabricados 2022.csv', index_col=0)
df1
```

```python
#Paso 2: Aplicamos filtro de filas vs columnas
# Aqui se filtran 2 valores del indice, los cuales son "F39835","F39837",
#de las cuales solo visualizamos la columna de nombre

filtro4=df1.loc[["F39835","F39837"], ["NOMBRE"]]
filtro4.to_csv("filtro4.csv")
files.download("filtro4.csv")
filtro4
```

```python
# Se filtran los primeros 5 registros del DF
filtro5=df.head()
filtro5.to_csv("filtro5.csv")
files.download("filtro5.csv")
filtro5
```

```python
#Se filtran los ultimos 5 registros del df
filtro6=df.tail()
filtro6.to_csv("filtro6.csv")
files.download("filtro6.csv")
filtro6
```

```python
# Se filtran tolos los valores de CANT que sean mayores a 8
filtro7=df[df["CANT"] > 8]
filtro7.to_csv("filtro7.csv")
files.download("filtro7.csv")
filtro7
```

```python
# Se filtran aquellos valores en donde CANT sea mayor a 15 y
# el nombre sea ALFREDO CANELA
filtro8=df[(df["CANT"] > 15) & (df["NOMBRE"] == "ALFREDO CANELA")]
filtro8.to_csv("filtro8.csv")
files.download("filtro8.csv")
filtro8
```

```python
# Se filtran los valores en donde CANT sea menor a 5 o CVE_ART sea BOBINA001
filtro9=df[(df["CANT"] < 5) | (df["CVE_ART"] == "BOBINA001")]
filtro9.to_csv("filtro9.csv")
files.download("filtro9.csv")
filtro9
```
